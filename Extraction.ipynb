```python
import pytesseract
pytesseract.pytesseract.tesseract_cmd=r'C:\Program Files\Tesseract-OCR\tesseract.exe'
import matplotlib.pyplot as plt
import cv2
import keras_ocr
%matplotlib inline
```

```python
url = [
     "https://i.redd.it/7wpmbsckae371.jpg",
     r"C:\Users\devap\Downloads\Sample-handwritten-text-input-for-OCR.png",
    "https://raw.githubusercontent.com/sanskar-hasija/ocr-comparision/main/test_images/image3.png",
    "http://test.ultralinx.co/wp-content/uploads/2020/03/1_beautiful-handwriting-lettering-calligraphy-58-572b35d8d1992__700jpg.jpg"]
images = [ keras_ocr.tools.read(i) for i in url]
fig = plt.figure(figsize=(16,10))
```

```python
fig = plt.figure(figsize=(16,10))
rows = 2
columns = 2

fig.add_subplot(rows, columns, 1)
plt.imshow(images[0])
plt.axis('off')
plt.title("First Image")

fig.add_subplot(rows, columns, 2)
plt.imshow(images[1])
plt.axis('off')
plt.title("Second Image")

fig.add_subplot(rows, columns, 3)
plt.imshow(images[2])
plt.axis('off')
plt.title("Third Image")

fig.add_subplot(rows, columns, 4)
plt.imshow(images[3])
plt.axis('off')
plt.title("Fourth Image");
```

```python
tesseract_preds = []
for img in images:
    tesseract_preds.append(pytesseract.image_to_string(img))
```

```python
for i in range(4):
    with open("Documents/text"+str(i+1)+".txt", 'w') as f:
        f.write(tesseract_preds[i]);
        print("File saved!")
```
